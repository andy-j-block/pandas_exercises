```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
df= pd.read_csv(url, index_col=0)
df.head()
```

### Step 4. Delete the Unnamed 0 column

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(20,10))
sns.distplot(df['total_bill'])
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
df['tip_pcnt']= (df['tip']/df['total_bill'])*100
sns.set_style('darkgrid')
plt.figure(figsize=(20,10))
sns.jointplot(data=df, x='total_bill', y='tip_pcnt',kind='reg')
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
by_size=df.groupby('size').agg({'total_bill':'mean','tip_pcnt':'mean'})
by_size
```

```python
plt.subplots(figsize=(20,10))
ax1=sns.barplot(data=by_size,x=by_size.index,y='total_bill',color='blue')
ax2=ax1.twinx()
ax2=sns.lineplot(data=by_size,x=by_size.index-1,y='tip_pcnt',color='orange')
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(15,10))
g=sns.FacetGrid(data=df, col='day')
g.map_dataframe(sns.histplot, x='total_bill')
```
